## Importing Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
import scipy.stats as s
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn import preprocessing
import math
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB 
from sklearn.naive_bayes import MultinomialNB 
from sklearn.naive_bayes import CategoricalNB
```

```python
#Loading the data
df=pd.read_csv("../input/digit-recognizer/train.csv")
df
```

```python
#Checking Columns
df.columns
```

```python
#Checking Unique values in Label columns
df["label"].value_counts()
```

```python
#Checking for null Values
df.isnull().sum()
```

```python
#Checking dataframe for any invalid/missing data
df.info()
```

```python
#Creating X & Y for Train_test_split
X = df.drop(["label"], axis=1)
y = df["label"]
```

```python
# Normalizing values
X = X.astype('float64') / 255
```

```python
#spliting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

### Linear Regression

```python
#creating and training the model
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
#predicting the splited data
y_test_pred_reg = reg.predict(X_test)
```

```python
acc=[]
#checking the accuracy
acc.append(metrics.r2_score(y_test, y_test_pred_reg)*100)
acc
```

## Naive Bayes

```python
#fitting and predicting using naive bayes
gnb = GaussianNB()
y_pred_gnb = gnb.fit(X_train, y_train).predict(X_test)
bnb = BernoulliNB()
y_pred_bnb = bnb.fit(X_train, y_train).predict(X_test)
mnb = MultinomialNB()
y_pred_mnb = mnb.fit(X_train, y_train).predict(X_test)
```

```python
#checking accuracy of naive bayes models
accuracy_gnb = metrics.accuracy_score(y_test, y_pred_gnb)
accuracy_bnb = metrics.accuracy_score(y_test, y_pred_bnb)
accuracy_mnb = metrics.accuracy_score(y_test, y_pred_mnb)
print("Accuracy of GaussianNB: ",accuracy_gnb," Accuracy of BernoulliNB: ",accuracy_bnb," Accuracy of MultinomialNB: ",accuracy_mnb)
acc.append(accuracy_bnb*100)#appending the top accuracy of naive bayes to compare
```

## SVM

```python
#creating and training the model
svm_clf = SVC(kernel="rbf", random_state=42, verbose=3,C=9)
svm_clf.fit(X_train, y_train)
```

```python
#predicting the splited data
y_test_pred_svm = svm_clf.predict(X_test)
```

```python
#checking accuracy of svm model
acc.append(metrics.accuracy_score(y_test, y_test_pred_svm)*100)
acc#almost 0.98 accuracy
```

```python
model = [ 'Regression', 'Naive Bayes', 'SVM']
accuracy=pd.DataFrame(zip(model, acc),columns=['Models','Accuracy'])
accuracy
```

## The best model based on accuracy is the Support Vector Classification (0.98)

## Cross Validation

```python
# ********* REG ********* 
reg_scores = cross_val_score(reg, X_train, y_train, cv=6)
print('Reg Scores: ',reg_scores)
```

```python
# ********* NAIVE BAYES ********* 
nb_scores = cross_val_score(bnb, X_train, y_train, cv=6)
print('Naive Bayes Scores: ',nb_scores)
```

```python
# ********* SVM ********* 
svc_scores = cross_val_score(svm_clf, X_train, y_train, cv=6)
print('SVC Scores: ',svc_scores)
```

```python
#MeanScores
meanScores=[]
meanScores.append(reg_scores.mean())
meanScores.append(nb_scores.mean())
meanScores.append(svc_scores.mean())
```

```python
#df of cross calidation scores
cv = {'Models': [  'Regression','Naive Bayes','SVC'], 
    'Cross Validation': meanScores}

cv_df = pd.DataFrame(cv)
cv_df
```
